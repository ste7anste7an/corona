```python
import numpy as np
def create_matrix(l):
    kl=1
    K=np.zeros((l,kl))
    M=np.zeros((l,l))
    for n in range(kl):
      for i in range(0,l):
        z=np.zeros((1,l))
        k=np.zeros((1,kl))
        k.itemset((0,n),1)
        if n==0:
            z.itemset((0,i),1)
        if n==0 and i==0:
            z.itemset((0,(i-1)%l),1)
        K[i]=K[i]+K[(i-1)%l]+k
        M[i]=M[i]+M[(i-1)%l]+z
#     print("key matrix")
#     print(K)
#     print("msg matrix")
#     print(M)
    return(M,K)
```

```python
Z=np.linalg.inv(M)
```

```python
def encrypt(key, msg):
    key = list(key)
    msg = [ord(i) for i in msg]
    for char_key in key:
        for i in range(len(msg)):
            if i == 0:
                tmp =  msg[i] + ord(char_key) + msg[-1]
            else:
                tmp = msg[i] + ord(char_key) + msg[i-1]
            while tmp > 255:
                tmp -= 256
            msg[i] = tmp
    return msg 
```
